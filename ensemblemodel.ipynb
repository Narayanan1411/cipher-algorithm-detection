```python
def hex_to_bytes(hex_string, max_len=256):
    byte_seq = bytes.fromhex(hex_string)
    byte_seq = byte_seq.ljust(max_len, b'\x00')[:max_len]
    return byte_seq
```

```python
max_len = 256
byte_sequences = np.array([np.frombuffer(hex_to_bytes(seq, max_len), dtype=np.uint8).reshape(16, 16, 1)
                           for seq in dataset['byte_sequence']])
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

def predict_symmetric(arr):
    label1 = ['AES-256', 'Blowfish', 'DES']
    label2=['DSA', 'ECC', 'RSA-2048', 'RSA-4096']
    label3=['ChaCha20', 'RC4', 'Salsa20']
    encoder1 = LabelEncoder()
    encoder1.fit(label1)
    encoder2 = LabelEncoder()
    encoder2.fit(label2)
    encoder3 = LabelEncoder()
    encoder3.fit(label3)
    predicted_probabilities1 = model1.predict(arr)
    predicted_probabilities2 = model2.predict(arr)
    predicted_probabilities3 = model3.predict(arr)
    predicted_probabilities4 = model4.predict(arr)
    predicted_probabilities5 = model5.predict(arr)
    predicted_probabilities6 = model6.predict(arr)


    predicted_labels_with_probabilities1 = {label: prob for label, prob in zip(encoder1.classes_, predicted_probabilities1[0])}
    predicted_labels_with_probabilities2 = {label: prob for label, prob in zip(encoder1.classes_, predicted_probabilities2[0])}
    predicted_labels_with_probabilities3 = {label: prob for label, prob in zip(encoder2.classes_, predicted_probabilities3[0])}
    predicted_labels_with_probabilities4 = {label: prob for label, prob in zip(encoder2.classes_, predicted_probabilities4[0])}
    predicted_labels_with_probabilities5 = {label: prob for label, prob in zip(encoder3.classes_, predicted_probabilities5[0])}
    predicted_labels_with_probabilities6 = {label: prob for label, prob in zip(encoder3.classes_, predicted_probabilities6[0])}

    return(predicted_labels_with_probabilities1,predicted_labels_with_probabilities2,predicted_labels_with_probabilities3,predicted_labels_with_probabilities4,predicted_labels_with_probabilities5,predicted_labels_with_probabilities6)

```

```python
pred_model_1=[]
pred_model_2=[]
pred_model_3=[]
pred_model_4=[]
pred_model_5=[]
pred_model_6=[]
for i in byte_sequences:
    arr=predict_symmetric(np.array([i]))
    pred_model_1.append(arr[0])
    pred_model_2.append(arr[1])
    pred_model_3.append(arr[2])
    pred_model_4.append(arr[3])
    pred_model_5.append(arr[4])
    pred_model_6.append(arr[5])
```

```python
def print_val_symmetric(pred_model_1_0, pred_model_2_0):
    def ensemble_prediction(pred_model_1_0, pred_model_2_0):

        combined_predictions = {}

        combined_predictions.update(pred_model_1_0)
        combined_predictions.update(pred_model_2_0)
        final_prediction = max(combined_predictions, key=combined_predictions.get)
        return final_prediction, combined_predictions
    final_algorithm, all_predictions = ensemble_prediction(pred_model_1_0, pred_model_2_0)
    print(f"All Predictions: {all_predictions}")
    print(f"Predicted Algorithm: {final_algorithm}, Confidence: {all_predictions[final_algorithm]}")
```

```python
def print_val_asymmetric(pred_model_1_0, pred_model_2_0):
    def ensemble_prediction(pred_model_1_0, pred_model_2_0):
        combined_predictions = {}

        combined_predictions.update(pred_model_1_0)
        combined_predictions.update(pred_model_2_0)
        final_prediction = max(combined_predictions, key=combined_predictions.get)

        return final_prediction, combined_predictions

    final_algorithm, all_predictions = ensemble_prediction(pred_model_1_0, pred_model_2_0)
    print(f"All Predictions: {all_predictions}")
    print(f"Predicted Algorithm: {final_algorithm}, Confidence: {all_predictions[final_algorithm]}")
```

```python
def print_val_stream(pred_model_1_0, pred_model_2_0):
    def ensemble_prediction(pred_model_1_0, pred_model_2_0):

        combined_predictions = {}


        combined_predictions.update(pred_model_1_0)
        combined_predictions.update(pred_model_2_0)
        final_prediction = max(combined_predictions, key=combined_predictions.get)


        return final_prediction, combined_predictions


    final_algorithm, all_predictions = ensemble_prediction(pred_model_1_0, pred_model_2_0)
    print(f"All Predictions: {all_predictions}")
    print(f"Predicted Algorithm: {final_algorithm}, Confidence: {all_predictions[final_algorithm]}")
```
